```python
## Preprocessing
#drop remarks column
#set index sno
#rename col
#convert amount to cr rs
#date col to datetime 
#drop na
```

```python
# since df['Remarks'] contains more null values--just drop it.
df['Remarks'].isna().sum()
```

```python
df.drop('Remarks',axis=1,inplace=True)
```

```python
df
```

```python
#set index sno
df.set_index('Sr No',inplace=True)
```

```python
df
```

```python
#rename col
df.rename(columns={
    'Date dd/mm/yyyy':'date',
    'Startup Name':'startup',
    'Industry Vertical':'vertical',
    'SubVertical':'subVertical',
    'City  Location':'City',
    'Investors Name':'Investors',
    'InvestmentnType':'round',
    'Amount in USD':'amount'
    },inplace=True)
```

```python
df
```

```python
#convert amount to cr rs
df['amount']=df['amount'].fillna('0')
df['amount']=df['amount'].fillna('0')
df['amount']=df['amount'].str.replace(',','')
df['amount']=df['amount'].str.replace('undisclosed','0')
df['amount']=df['amount'].str.replace('unknown','0')
df['amount']=df['amount'].str.replace('Undisclosed','0')
df['amount']=df['amount'].fillna('0')
df['amount']=df['amount'].str.replace('Undisclosed','0')
df['amount']=df['amount'].str.replace('14342000+','143420000')
df['amount']=df['amount'].str.replace('\\\\xc2\\\\xa020000000','0')
df['amount']=df['amount'].str.replace('\\\\xc2\\\\xa016200000','0')
df['amount']=df['amount'].str.replace('\\xc2\\xa0600000','0')
df['amount']=df['amount'].str.replace('\\xc2\\xa0685000','0')
df['amount']=df['amount'].str.replace('\\xc2\\xa019350000','0')
df['amount']=df['amount'].str.replace('\\xc2\\xa05000000','0')
df['amount']=df['amount'].str.replace('\\xc2\\xa010000000','0')
```

```python
# if null values are there then then it is counted as float

df['amount']=df['amount'].fillna('0')
df['amount']=df['amount'].fillna('0')
df['amount']=df['amount'].str.replace(',','')
df['amount']=df['amount'].str.replace('undisclosed','0')
df['amount']=df['amount'].str.replace('unknown','0')
df['amount']=df['amount'].str.replace('Undisclosed','0')



def isdigit(a):
    if a.isdigit()==False:
        return '0'
    else:
        return a
    
df['amount']=df['amount'].apply(isdigit)


```

```python
df[~df['amount'].str.isdigit()]
```

```python
df
```

```python
df['amount']=df['amount'].astype('int64')
```

```python
df['amount']=(df['amount']*83)/10000000
df['amount']
```

```python
df
```

```python
#date col to datetime 
df['date']=pd.to_datetime(df['date'],errors='coerce')
```

```python
df['date']
```

```python
#advantage of datetime type
df['date'].dt.year
```

```python
#drop na
df=df.dropna(subset=['date','startup','vertical','City','Investors','round','amount'])
df
```
